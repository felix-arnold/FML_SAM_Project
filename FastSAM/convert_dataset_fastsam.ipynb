```python
from pycocotools.coco import COCO
from pycocotools import mask as maskUtils
import numpy as np
import json
from PIL import Image
import supervision as sv
import matplotlib.pyplot as plt 

import numpy as np
import matplotlib.pyplot as plt 

#datasets = ["Cat"]
datasets = ["Baseball bat", "Bird", "Breast", "Cat", "Chalk group", "Clock", "Cow", "Dog", "Dolphin above", "Dolphin below", "Pizza", "Stop sign", "Tie", "Salt dome", "Skin"]
```

```python
data_id = 0
for dataset in datasets:
    images = np.load('./datasets_bis/' + dataset + '/samples.npy', allow_pickle=True)
    labels = np.load('./datasets_bis/' + dataset + '/labels.npy', allow_pickle=True)

    i = 0
    for image in images:
        if i==321 and dataset == "Breast":
            i+=1
            continue
            
        image = Image.fromarray(image.astype('uint8'))
        number_str = '{:0>6}'.format(i + data_id * 400)
            
        if i < 0.8 * len(images):
            file_path = "datasets/fastsam_custom_small/images/train/" + number_str + ".png"
        else:
            file_path = "datasets/fastsam_custom_small/images/val/" + number_str + ".png"
        
        print(file_path)
        image.save(file_path)
        
        i+=1
    
    j = 0
    for label in labels:
        if j==321 and dataset == "Breast":
            j+=1
            continue
            
        width = len(label[0])
        height = len(label)
        first = True
        seg_data = ""
        
        polygons = sv.mask_to_polygons(labels[j])
    
        for polygon in polygons:
            if first:
                seg_data += str(data_id) + " "
                first = False
            else:
                seg_data += "\n" + str(data_id) + " "
            for point in polygon:
                seg_data += str(point[0]/width) + " " + str(point[1]/height) + " "
                
        number_str = '{:0>6}'.format(j + data_id * 400)
        
        if j < 0.8 * len(labels):
            file_path = "datasets/fastsam_custom_small/labels/train/" + number_str + ".txt"
        else:
            file_path = "datasets/fastsam_custom_small/labels/val/" + number_str + ".txt"

        print(file_path)
        with open(file_path, "w") as file:
            file.write(seg_data)
        
        j+=1
    
    data_id+=1
```
